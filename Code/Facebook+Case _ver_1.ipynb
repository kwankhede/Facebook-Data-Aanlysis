```r
p1<-ggplot(data= pf,aes(x=friend_count))+geom_histogram(color="black",fill="firebrick1")+ylab('Number of users in sample')+xlab('Number of facebook friends')+ggtitle("Original Scale")+theme(plot.title = element_text(hjust = 0.5))
p2<-p1+scale_x_log10()+geom_histogram(color="black",fill='firebrick1') +ylab('Number of users in sample')+ggtitle("log10()")+theme(plot.title = element_text(hjust = 0.5))
p3<-p1+scale_x_sqrt()+geom_histogram(color="black",fill='firebrick2')+ylab('Number of users in sample')+ggtitle("Square root")+theme(plot.title = element_text(hjust = 0.5))
grid.arrange(p1,p2,p3,ncol=1)
```

# Conditional Mean by using dplyr

```r
install.packages('dplyr',repos = "http://cran.us.r-project.org")
```

```r
library(dplyr)  # it is use to filter group,arrange and mutate 
```

```r
#group our dataset by age 
Age_groups<-group_by(pf,age)
```

```r
#Conditional mea
pf.fc.by.age<-summarise(Age_groups,friend_count_mean= mean(friend_count),friend_count_median= median(friend_count),n=n())
```

```r
head(pf.fc.by.age)
```

```r
pf.fc.by.age<-arrange(pf.fc.by.age,age)
```

```r
head(pf.fc.by.age)
```

# conditional mean by using alternate method

```r
pf_fc_by_age<-pf  %>% 
group_by(age) %>% 
summarise(Age_groups,friend_count_mean= mean(friend_count),friend_count_median= median(friend_count),n=n()) %>% 
arrange(age)
head(pf_fc_by_age,10)
```
