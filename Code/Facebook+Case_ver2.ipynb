# I. Facebook users according to their ages


As we know, Facebook is very popular in youngster. We also have strong that, most of the Facebook users are between 15-35 years old. It is our rough guess. I want to use my dataset to see people from which age group are mostly using Facebook. For that I am going to plot three different plots. One is histogram and two are scattered plot.

Let's see what our graphs tell us about our age group prediction for the most Facebook group user's age group.


```r
# Graph 1: 
#Number of users with different ages. 
ggplot(aes(x = age), data = pf) + geom_histogram(binwidth = 5, fill = 'deepskyblue3',color="black") + scale_x_continuous(breaks = seq(0, 113, 5))+ylab('Number of Facebook users')+
ggtitle("Age vs Numbers of users")+theme(plot.title = element_text(hjust = 0.5))+xlab('Age of Facebook users')
```

If we ask question about, which age group are using more Facebook? To answer this question, I plotted histogram of age vs number of Facebook users. From the above graph, we can see that, people between 17 to 27 years are active users of Facebook. 
There are some users more than 100 and more than 100 years old. 

This shows that there are some fake accounts or may be people used fake date of birth for their Facebook account



To take a detail look at my analysis "facebook users according to their ages", I created scatter plot. 

```r
# Graph 2:
ggplot(aes(x = age, y = friend_count), data = pf) + geom_jitter(color='deepskyblue3')+xlim(13,90) +ylab('Number of users in sample')+
ggtitle("Age vs Numbers of users")+theme(plot.title = element_text(hjust = 0.5))+xlab('Age of Facebook users')
```

To see details about Facebook user's age, I plotted above graph. From above graph we can see that, there are some outliers in our dataset. We must take care of outliers while doing data analysis.

Let’s run summary statistics to check minimum and maximum age of Facebook users to update of x axis scale


```r
summary(pf$age)
```

```r
#Graph 3:
#Trim X axis according to result from summary statistics.Min age of facebook user is 13 year and max is 113, 
#But we are going to remove entries whoes ages more than 90 years.
ggplot(aes(x = age, y = friend_count), data = pf) + geom_point(color="deepskyblue3")+xlim(13,90) +ylab('Number of users in sample')+
ggtitle("Age vs Numbers of users")+theme(plot.title = element_text(hjust = 0.5))+xlab('Age of Facebook users')
```

From above three graphs we can say that, people between 17 to 27 ages are using Facebook more than the other group. So, our initial guess was quite right. With the help of visualization and running summary statistics, Facebook data analysis is showing characteristics of our data.  

# II.  Relation of Facebook friends with gender

In above section about number of facebook friends by user's age.Next, I wantto see a relation between number of facebook friends and gender.
Here again I am going to create three dieffrent graph and fro that I will try to analyze relation between facebook friends count with gender.

```r
#Graph 1
#Total friends on facebook by gender
ggplot(aes(x = friend_count), data = pf) + geom_histogram(fill="deepskyblue3",color='black') + scale_x_continuous(limits = c(0, 1000), breaks = seq(0, 1000, 50)) + 
facet_wrap(~gender)+ggtitle('Friends count by gender')+xlab("Friends count")+ylab('Number of users in sample')+theme(plot.title = element_text(hjust = 0.5))
```

From above graph we can see Males have more facebook friends than females.In our dataset we can see NA entries. NA entries are those entreis where gender is not disclosed. 
Lets remove those entries and make a new graph without NA entries 

```r
#Graph 2
#Removing NA entries from the dataset 
#data = subset(pf, !is.na(gender)
ggplot(aes(x = friend_count), data = subset(pf, !is.na(gender))) + geom_histogram(fill="deepskyblue3",color='black') + scale_x_continuous(limits = c(0, 1000), breaks = seq(0, 1000, 50)) + facet_wrap(~gender)+ggtitle('Freinds count and Frequency')+xlab("Friends count")+ylab('Number of users in sample')+theme(plot.title = element_text(hjust = 0.5))
```

We can’t say easily from above plotted graphs about which gender has more number of Facebook friends.
To support our claim, next I am going to create line graph. With updated y axis.
In the next graph, I have updated Y axis i.e. Facebook users by in percentage. (count/sum(count))

```r
#Graph 3
ggplot(aes(x = friend_count, y = ..count../sum(..count..)), data = subset(pf, !is.na(gender))) + geom_freqpoly(aes(color = gender), binwidth=10) + 
scale_x_continuous(limits = c(0, 1000), breaks = seq(0, 1000, 50)) + xlab('Friend Count') + ylab('Percentage of users with that friend count')+ggtitle("Friend counts by Gender")+theme(plot.title = element_text(hjust = 0.5))
```

In above graph we can clearly see that facebook friends for males are highe than females.  

We can check the same thing by using "by" function in R.

```r
table(pf$gender)
by(pf$www_likes,pf$gender,sum)
```

It is clear now that male facebook users have more facebook friends than female .

```r
#Average friends count by gender 
by(pf$friend_count,pf$gender,summary)   # pf$gender is categorical variable , we are trying to get statistics by gender 
```

From above summary we can say that avaerage friend count for female are more than male.

# III. Number of Facebook friends and Facebook active years of a user

In this section I am going to check the realtion between number of facebook users with the time they have been using facebook.
Lets plot a graph of number of users with facebook user's tenure.

```r
ggplot(aes(x = (tenure / 365),), data = pf) + geom_histogram(color = 'black', fill = "deepskyblue3") + scale_x_continuous(breaks = seq(1, 7, 1), limits = c(0, 7)) + xlab('Number of years using Facebook') + ylab('Number of users in sample')+ggtitle("Years vs Users")+theme(plot.title = element_text(hjust = 0.5))
```

```r
ggplot(aes(x = (tenure / 365), y= pf$friend_count), data = pf) + geom_point(alpha=1/20,color  = "deepskyblue3") + xlab('Number of years using Facebook') +xlim(0,7)+ ylab('Number of users in sample')+ggtitle("Years vs Users")+theme(plot.title = element_text(hjust = 0.5))
```

From above graph we can see that most of the facebook users have been using facebook from last 2-3 years.

# IV. Now lets see the date of birth of Facebook users. 

As a practice of an exposatory data analysis, I tried to do some analysis on date of birth of facebook users. In my dataset we have day of date of birth. I tried to plot a graph of facebook user's day of birth.

```r
ggplot(aes(x = dob_day), data = pf) + geom_histogram(binwidth = 1,color ="black",fill ="deepskyblue3") + scale_x_continuous(breaks = 1:31)+ xlab("Day of birth")+ylab('Number of users in sample') +ggtitle("Facebook user's day of bitrth")+theme(plot.title = element_text(hjust = 0.5))
```

From above graph we can infer that,there are many facebook users whose day of birth is 1 st in every month. It shows there are some fake account holders which just select autodate option while sign up into the facebook. Morever we can see on 31st ther are fewer entries than the other days, it is right because there is no 31st in every month.

```r
 ggplot(aes(x = dob_day), data = pf) + geom_histogram(binwidth = 1,color ="black",fill ="deepskyblue3") + scale_x_continuous(breaks = 1:31)+ xlab("Day of birth")+ylab('Number of users in sample') +ggtitle("Facebook user's day of bitrth")+facet_wrap(~dob_month,ncol=3)+theme(plot.title = element_text(hjust = 0.5))
```
